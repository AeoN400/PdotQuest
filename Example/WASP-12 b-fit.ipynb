```python
from PdotQuest.readata import Read_Data
from PdotQuest.pdotfit import *
```

```python
rd = Read_Data('WASP-012')
tepoch,ttime,terror = rd.tess_pytransit()
```

```python
p_0 = 1.091423
```

```python
fig=plot_resi(tepoch,ttime,terror)
```

# 1. Linear Fit

```python
linfit = LinearFit(ttime,terror,p_0,rej_sigma=5)
```

```python
linfit.run_MCMC0()
```

```python
linfit.BIC0()
```

```python
linfit.plot_residuals0()
```

# 2. Quadratic Fit

```python
qfit = QuadraFit(ttime,terror,p_0,[-40,-20],rej_sigma=5)
```

```python
qfit.run_MCMC()
```

```python
qfit.plot_residuals(tess=True)
```

```python
qfit.BICq()
```

```python
qfit.qorner()
```

```python
qfit.print_posterior()
```
